# Bifurcation diagram

```python
import numpy as np
import matplotlib.pyplot as plt
from system import System

a = 1.5
m = 0.45
u0 = lambda x, y: a
v0 = lambda x, y: np.random.rand()

S = System(100, 100, 0.5, 0.05, 10, 0.1, m, a, u0, v0)
fig, ax = plt.subplots(figsize=(8, 8))
S.bifurcation_diagram(0.05, 50, log=True, ax=ax)

aa = np.linspace(0.9, 1.5, 100)
plus = (aa + np.sqrt(aa**2 - 4 * m**2))/(2 * m)
minus = (aa - np.sqrt(aa**2 - 4 * m**2))/(2 * m)
ax.plot(aa, plus, color="green", linestyle="dashed", label=r"$\frac{a^2 + \sqrt{a^2 - 4 m^2}}{2m}$")
ax.plot(aa, minus, color="green", linestyle="dotted", label=r"$\frac{a^2 - \sqrt{a^2 - 4 m^2}}{2m}$")

ax.legend()

fig.savefig("bifurcation.png", transparent=True)
plt.show()
```
