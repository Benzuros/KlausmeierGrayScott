```python
import sys
from pathlib import Path

project_root = Path.cwd().parents[0]
sys.path.append(str(project_root / "pipeline"))

import numpy as np
import matplotlib.pyplot as plt
from system import System

xmax = ymax = 100
h = 0.5
ht = 0.05
d1 = 10
d2 = 0.1
v0 = lambda x, y: np.random.rand()
```

# Some examples of pattern generation

## Changing *a*

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(8, 8), layout="constrained")

m = 0.45

for ax, a in (ax1, 1.15), (ax2, 1.25), (ax3, 1.35), (ax4, 1.45):
    s = System(xmax, ymax, h, ht, d1, d2, m, a, lambda x, y: a, v0)
    s.simulate_until_stable(50, extend=False)
    s.plot(fig=fig, ax=ax)

fig.savefig("zoo_a.png", transparent=True)
plt.show()
```

## Changing *m*

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(8, 8), layout="constrained")

a = 1.25

for ax, m in (ax1, 0.4), (ax2, 0.45), (ax3, 0.5), (ax4, 0.55):
    s = System(xmax, ymax, h, ht, d1, d2, m, a, lambda x, y: a, v0)
    s.simulate_until_stable(50, extend=False)
    s.plot(fig=fig, ax=ax)

fig.savefig("zoo_m.png", transparent=True)
plt.show()
```
