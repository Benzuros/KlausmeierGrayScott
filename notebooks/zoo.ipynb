```python
import sys
from pathlib import Path

project_root = Path.cwd().parents[0]
sys.path.append(str(project_root / "pipeline"))

import numpy as np
import matplotlib.pyplot as plt
from system import System


v0 = lambda x, y: np.random.rand()
```

# Error

```python
Fig, ((Ax0, Ax1), (Ax2, Ax3)) = plt.subplots(ncols=2, nrows=2)
s = System(100, 100, 0.5, 0.05, 10, 0.1, 0.45, 1.25, lambda x, y: 1.25, v0)

print(s.error(100, until_stable=True, fig=Fig, ax=(Ax0, Ax1, Ax2, Ax3), tolerance=0.01))
plt.show()
```
